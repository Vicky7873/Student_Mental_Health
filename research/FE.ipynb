```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('MentalHealthSurvey.csv')
```

```python
df.head()
```

```python
cat_cols=[]
num_cols=[]
for col in df.columns:
    if df[col].dtype=='object':
        cat_cols.append(col)
    else:
        num_cols.append(col)
```

```python
df.head(2)
```

```python
df['gender'] = df['gender'].apply(lambda x: 1 if x == 'Male' else 0)
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder,OneHotEncoder
```

```python
from sklearn.preprocessing import OneHotEncoder
le = LabelEncoder()
ohe = OneHotEncoder()
oe = OrdinalEncoder()
```

```python
cat_cols.remove('gender')
```

```python
cat_cols
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from scipy.sparse import issparse

# Create OHE and LE instances
ohe = OneHotEncoder()
le = LabelEncoder()

# Iterate over categorical columns
for col in df[cat_cols]:
    # Check if column is in the list of columns to be OHE
    if col in ['university', 'degree_major', 'residential_status']:
        # Fit and transform the column using OHE
        encoded_col = ohe.fit_transform(df[[col]])
        if issparse(encoded_col):
            # Convert to dense array if it's sparse
            encoded_col = encoded_col.toarray()
        df[col] = encoded_col
    # Check if column is "campus_discrimination"
    elif col == "campus_discrimination":
        # Fit and transform the column using LE
        df[col] = le.fit_transform(df[[col]])
    # For other categorical columns
    else:
        # Fit and transform the column using OHE (or LE if appropriate)
        encoded_col = oe.fit_transform(df[[col]])
        if issparse(encoded_col):
            # Convert to dense array if it's sparse
            encoded_col = encoded_col.toarray()
        df[col] = encoded_col
```
